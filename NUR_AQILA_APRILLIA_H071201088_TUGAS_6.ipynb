**2. DATA TEXT**

* cara 1

Fungsi Tokenizer pada text :

Memisah kata-kata / dikelompokkan yang awalnya menjadi satu atau kalimat. 

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

num_words : menentukan jumlah kata 
oov_token : Untuk mengganti kata yang sulit di deteksi 

```python
tokenizer = Tokenizer(num_words=20, oov_token="-")
```

```python
teks = ["saya sangat suka belajar data mining",
        "Data mining sangat menarik"]
```

```python
tokenizer.fit_on_texts(teks)
```

penyimpanan di sequence

```python
print(_)
```

```python
seq = tokenizer.texts_to_sequences(teks)
```

```python
print(tokenizer.word_index)
```

**Menyeragamkan hasil tokenizer **

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
sequence_sama_panjang = pad_sequences(seq)
```

```python
print(sequence_sama_panjang)
```
